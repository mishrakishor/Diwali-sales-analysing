```python
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
####change datatype of amount
df["Amount"]=df["Amount"].astype('int')
```

data analysis

```python
####Gender count for shooping
gs=sns.countplot(x='Gender',data=df)
for bar in gs.containers:
    gs.bar_label(bar)
```

```python
#### who are big amount buyer compare of male and female
sales_amount=df.groupby(["Gender"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
b=sns.barplot(x='Gender',y='Amount',data=sales_amount)
for bar in b.containers:
    b.bar_label(bar)
```

```python
#####find many more time shopping by age group
agrp=sns.countplot(data=df,x="Age Group", hue= "Gender")
for bars in agrp.containers:
    agrp.bar_label(bars)
```

```python
##total amount vs age group
sales_amount=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Age Group",y="Amount",data=sales_amount)
```

```python
df.columns
```

```python
##### total number order of state top 10
sale_state=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(18,5)})
b=sns.barplot(x="State",y="Orders",data=sale_state)
for bar in b.containers:
    b.bar_label(bar)
```

```python
######total amount sales of sate top 10
sale_state=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(18,5)})
b=sns.barplot(x="State",y="Amount",data=sale_state)
for bar in b.containers:
    b.bar_label(bar)
```

```python
ms=sns.countplot(data=df,x="Marital_Status")
sns.set(rc={"figure.figsize":(8,5)})
for bar in ms.containers:
    ms.bar_label(bar)
```

```python
######find marital status with gender vs amount
marital_gen=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(5,5)})
result=sns.barplot(data=marital_gen,x="Marital_Status",y="Amount",hue="Gender")
```
